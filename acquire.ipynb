```python
import pandas as pd
import requests
import math
```

```python
people = []
for page in range(1,10):
    url = f'https://swapi.dev/api/people/{page}'
    response = requests.get(url).json()
    people.append(response)
pd.DataFrame(people).to_csv('people.csv')
```

```python
planets = []
for page in range(1,8):
    url = f'https://swapi.dev/api/planets/{page}'
    response = requests.get(url).json()
    planets.append(response)
pd.DataFrame(planets).to_csv('planets.csv')
```

```python
starships = []
for page in range(1,5):
    url = f'https://swapi.dev/api/starships/{page}'
    response = requests.get(url).json()
    starships.append(response)
pd.DataFrame(starships).to_csv('starships.csv')
```

```python
url = 'https://swapi.dev/api/starships' #Make initial call to get 'count' - know how
response = requests.get(url).json()
pages = math.ceil(response['count']*.1)

l = []

for p in range(1,pages+1):
    url = f'https://swapi.dev/api/starships/?page={p}'
    response = requests.get(url).json()
    for ship in response['results']:
        ship['url'] = url
        l.append(ship)
```

```python
pd.DataFrame(l).to_csv('starships.csv')
starships = pd.read_csv('starships.csv',index_col=0)
starships.head()
```
